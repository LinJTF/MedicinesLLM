```python
def create_boxplot(list_df, list_title):
    for i in range(len(list_df)):
        fig, ax = plt.subplots(1, 2, figsize=(16, 6))

        colors = ["skyblue", "lightcoral"]

        sns.boxplot(data=list_df[i]["Bleu Score"], ax=ax[0], color=colors[0])
        ax[0].set_title("Bleu Score Distribution", fontsize=14)
        ax[0].set_xlabel("Bleu Score", fontsize=12)
        ax[0].grid(axis="y", linestyle="--", alpha=0.7)

        sns.boxplot(data=list_df[i]["Rouge Score"], ax=ax[1], color=colors[1])
        ax[1].set_title("Rouge Score Distribution", fontsize=14)
        ax[1].set_xlabel("Rouge Score", fontsize=12)
        ax[1].grid(axis="y", linestyle="--", alpha=0.7)

        fig.suptitle(list_title[i], fontsize=16)

        sns.set_theme(style="darkgrid")

        plt.tight_layout()
        plt.show()

create_boxplot(list_df, list_title)
```

```python
list_df = [df_falcon1, df_falcon2, df_llama_7b_1, df_llama_7b_2, df_llama_13b_1, df_llama_13b_2, df_gpt1, df_gpt2]
list_title = ["Falcon-7b OpenAI Embedding", "Falcon-7b HuggingFace Embedding", " LLaMA2-7b  OpenAI Embedding", "LLaMA2-7b HuggingFace Embedding", "LLaMA2-13b OpenAI Embedding", "LLaMA2-13b HuggingFace Embedding", "GPT-3.5-Turbo OpenAI Embedding", "GPT-3.5-Turbo HuggingFace Embedding"]

def create_boxplot_all(list_df, list_title, score_type):
    fig, ax = plt.subplots(figsize=(18, 8))

    colors = ["skyblue", "lightcoral"]
    labels = ["OpenAI Embedding", "Multi-QA-Mpnet-Base-Dot-V1 Embedding"]

    boxplot_data = []
    boxplot_labels = []

    for i in range(len(list_df)):
        if score_type == "Bleu":
            boxplot_data.append(list_df[i]["Bleu Score"])
            boxplot_labels.append(list_title[i])
            ax.set_title("Bleu Score Distribution", fontsize=16)
            ax.set_xlabel("Bleu Score", fontsize=14)
        elif score_type == "Rouge":
            boxplot_data.append(list_df[i]["Rouge Score"])
            boxplot_labels.append(list_title[i])
            ax.set_title("Rouge Score Distribution", fontsize=16)
            ax.set_xlabel("Rouge Score", fontsize=14)

    sns.boxplot(data=boxplot_data, ax=ax, palette=colors)
    ax.set_xticklabels(boxplot_labels, rotation=45, ha='right', fontsize=12)
    ax.grid(axis="y", linestyle="--", alpha=0.7)

    sns.set_theme(style="darkgrid")

    legend_handles = [Patch(color=color, label=label) for color, label in zip(colors, labels)]

    legend = plt.legend(handles=legend_handles, title="Embedding", loc='upper right', fontsize=12)
    legend.get_title().set_fontsize(14)

    plt.tight_layout()
    plt.show()

create_boxplot_all(list_df, list_title, score_type="Bleu")
create_boxplot_all(list_df, list_title, score_type="Rouge")
```
