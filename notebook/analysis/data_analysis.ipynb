```python
def create_boxplot(list_df, list_title):
    for i in range(len(list_df)):
        fig, ax = plt.subplots(1, 2, figsize=(16, 6))

        colors = ["skyblue", "lightcoral"]

        sns.boxplot(data=list_df[i]["Bleu Score"], ax=ax[0], color=colors[0])
        ax[0].set_title("Bleu Score Distribution", fontsize=14)
        ax[0].set_xlabel("Bleu Score", fontsize=12)
        ax[0].grid(axis="y", linestyle="--", alpha=0.7)

        sns.boxplot(data=list_df[i]["Rouge Score"], ax=ax[1], color=colors[1])
        ax[1].set_title("Rouge Score Distribution", fontsize=14)
        ax[1].set_xlabel("Rouge Score", fontsize=12)
        ax[1].grid(axis="y", linestyle="--", alpha=0.7)

        fig.suptitle(list_title[i], fontsize=16)

        plt.tight_layout()
        plt.show()

create_boxplot(list_df, list_title)
```
